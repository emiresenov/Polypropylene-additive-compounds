```python
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d

# read your two-full-header CSVs as float arrays
eps_df = pd.read_csv('eps_p_full.csv', skiprows=2, header=None)
tan_df = pd.read_csv('tan_d_full.csv', skiprows=2, header=None)

# common log-grid
w = np.logspace(-1, 4, 100)

concs = [3, 1, 0.5, 0]
pairs = [(0,1), (2,3), (4,5), (6,7)]
blocks = []

for (cf, cv), c in zip(pairs, concs):
    # drop any NaNs in the raw columns before interp
    df_eps = eps_df[[cf, cv]].dropna()
    df_tan = tan_df[[cf, cv]].dropna()
    freq_eps, eps_raw = df_eps[cf].values, df_eps[cv].values
    freq_tan, tan_raw = df_tan[cf].values, df_tan[cv].values

    # build interpolators (linear in log10, extrapolate beyond ends)
    interp_eps = interp1d(np.log10(freq_eps), eps_raw,
                          kind='linear', fill_value='extrapolate')
    interp_tan = interp1d(np.log10(freq_tan), tan_raw,
                          kind='linear', fill_value='extrapolate')

    # sample them on your grid
    eps_i = interp_eps(np.log10(w))
    tan_i = interp_tan(np.log10(w))

    # **CLIP** (not nan) any negative values
    eps_i = np.clip(eps_i, 0, None)
    tan_i = np.clip(tan_i, 0, None)

    blocks.append(pd.DataFrame({
        'w'    : w,
        "eps'": eps_i,
        'tand' : tan_i,
        'conc' : c
    }))

df_all = pd.concat(blocks, ignore_index=True)
df_all.to_csv('PP_OH_all_concs.csv', index=False)
print("Done — no missing values any more!") 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1) Read the combined CSV
df = pd.read_csv('PP_OH_all_concs.csv')

# 2) Get the list of unique concentrations
concs = sorted(df['conc'].unique())

# 3) Plot ε′(ω) for each concentration
plt.figure(figsize=(8,5))
for c in concs:
    sub = df[df['conc']==c]
    plt.semilogx(sub['w'], sub["eps'"], marker='o', linestyle='-',
                 label=f"conc = {c}")
plt.xlabel('ω (rad/s)')
plt.ylabel("ε′")
plt.title("Dielectric Constant vs Frequency")
plt.legend()
plt.grid(True, which='both', ls='--', alpha=0.5)
plt.tight_layout()

# 4) Plot tan δ(ω) for each concentration
plt.figure(figsize=(8,5))
for c in concs:
    sub = df[df['conc']==c]
    plt.semilogx(sub['w'], sub['tand'], marker='o', linestyle='-',
                 label=f"conc = {c}")
plt.xlabel('ω (rad/s)')
plt.ylabel("tan δ")
plt.title("Loss Tangent vs Frequency")
plt.legend()
plt.grid(True, which='both', ls='--', alpha=0.5)
plt.tight_layout()

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.signal import savgol_filter

# Load the CSVs
eps_df = pd.read_csv('eps_p_full.csv', skiprows=2, header=None)
tan_df = pd.read_csv('tan_d_full.csv', skiprows=2, header=None)

# Frequency grid
w = np.logspace(-1, 4, 100)
logw = np.log10(w)

concs = [3, 1, 0.5, 0]
pairs = [(0,1), (2,3), (4,5), (6,7)]
blocks = []

for (cf, cv), c in zip(pairs, concs):
    # Drop NaNs
    df_eps = eps_df[[cf, cv]].dropna()
    df_tan = tan_df[[cf, cv]].dropna()
    freq_eps, eps_raw = df_eps[cf].values, df_eps[cv].values
    freq_tan, tan_raw = df_tan[cf].values, df_tan[cv].values

    # Interpolate in log-frequency space
    interp_eps = interp1d(np.log10(freq_eps), eps_raw, kind='linear', fill_value='extrapolate')
    interp_tan = interp1d(np.log10(freq_tan), tan_raw, kind='linear', fill_value='extrapolate')
    eps_i = interp_eps(logw)
    tan_i = interp_tan(logw)

    # Clip negatives
    eps_i = np.clip(eps_i, 0, None)
    tan_i = np.clip(tan_i, 0, None)

    # Apply Savitzky-Golay smoothing
    window_length = 15  # Must be odd, < len(w)
    polyorder = 2
    eps_i = savgol_filter(eps_i, window_length, polyorder)
    tan_i = savgol_filter(tan_i, window_length, polyorder)

    # Store results
    blocks.append(pd.DataFrame({
        'w': w,
        "eps'": eps_i,
        'tand': tan_i,
        'conc': c
    }))

# Combine and save
df_all = pd.concat(blocks, ignore_index=True)
df_all.to_csv('PP_OH_all_concs_savgol.csv', index=False)
print("Done — smoothed using Savitzky–Golay filter!")

# === Plotting ===

concs = sorted(df_all['conc'].unique())

# Dielectric Constant Plot
plt.figure(figsize=(8, 5))
for c in concs:
    sub = df_all[df_all['conc'] == c]
    plt.semilogx(sub['w'], sub["eps'"], marker='o', linestyle='-', label=f"conc = {c}")
plt.xlabel('ω (rad/s)')
plt.ylabel("ε′")
plt.title("Dielectric Constant vs Frequency (Savitzky–Golay Smoothed)")
plt.legend()
plt.grid(True, which='both', ls='--', alpha=0.5)
plt.tight_layout()

# Loss Tangent Plot
plt.figure(figsize=(8, 5))
for c in concs:
    sub = df_all[df_all['conc'] == c]
    plt.semilogx(sub['w'], sub['tand'], marker='o', linestyle='-', label=f"conc = {c}")
plt.xlabel('ω (rad/s)')
plt.ylabel("tan δ")
plt.title("Loss Tangent vs Frequency (Savitzky–Golay Smoothed)")
plt.legend()
plt.grid(True, which='both', ls='--', alpha=0.5)
plt.tight_layout()

plt.show()
```
